# Руководство по созданию Pull Request в GitHub

## Основные шаги

### 1. Подготовка репозитория
- **Форкните** (ответвите) репозиторий, нажав кнопку "Fork" в правом верхнем углу страницы репозитория
- **Клонируйте** свой форк на локальную машину:
  ```bash
  git clone https://github.com/ваш-логин/название-репозитория.git
  ```

### 2. Создание ветки для изменений
- Создайте новую ветку для ваших изменений:
  ```bash
  git checkout -b название-ветки
  ```
  Название ветки должно отражать суть изменений (например: `fix_login_bug` или `add_dark_theme`)

### 3. Внесение изменений
- Внесите необходимые изменения в код
- Проверьте, что изменения работают корректно
- Добавьте изменения в отслеживаемые файлы:
  ```bash
  git add .
  ```

### 4. Коммит изменений
- Создайте коммит с понятным описанием:
  ```bash
  git commit -m "Краткое описание изменений"
  ```
  Хорошее описание коммита:
  - Начинается с глагола (Исправил, Добавил, Улучшил)
  - Лаконично описывает суть изменений

### 5. Отправка изменений на GitHub
- Запушьте вашу ветку в ваш форк:
  ```bash
  git push origin название-ветки
  ```

### 6. Создание Pull Request
1. Перейдите на страницу вашего форка на GitHub
2. Нажмите кнопку "Compare & pull request"
3. Заполните форму:
   - **Title** - краткое название ваших изменений
   - **Description** - подробное описание:
     - Что было изменено
     - Почему эти изменения важны
     - Какие проблемы решают
4. Укажите связанные issues (если есть), используя ключевые слова (например: "Closes #123")
5. Нажмите "Create pull request"

## Рекомендации
- Перед созданием PR убедитесь, что ваш код соответствует стилю проекта
- Проверьте, что все тесты проходят
- Обновите документацию, если это необходимо
- Будьте готовы к обсуждению и доработке вашего кода

## Шаблон описания Pull Request
```markdown
## Описание изменений
[Кратко опишите, что было изменено]

## Причина изменений
[Объясните, почему эти изменения необходимы]

## Тестирование
[Опишите, как вы протестировали изменения]

## Скриншоты (если применимо)
[Добавьте скриншоты, если изменения визуальные]

## Связанные Issues
[Укажите номера связанных issues, например: Closes #123]
```
